axis=0 means travel vertically (down across the rows).
axis=1 means travel horizontally (left-to-right across the columns).

```python
import pandas as pd

# 1. The Setup: Creating our tiny test score DataFrame
data = {
    'Math (A)': [10, 30, 50], 
    'Science (B)': [20, 40, 60]
}

# We use 'index' to name our rows so it matches the example
df = pd.DataFrame(data, index=['Student 0', 'Student 1', 'Student 2'])

print("=== ORIGINAL DATAFRAME ===")
print(df)
print("\n" + "="*30 + "\n")




```

```python
# 2. Scenario 1: axis=0
print("=== SUM WITH axis=0 ===")
print("Traveling VERTICALLY down the rows. Expect column totals:")
print(df.sum(axis=0))
print("\n" + "="*30 + "\n")
```

```python
# 3. Scenario 2: axis=1
print("=== SUM WITH axis=1 ===")
print("Traveling HORIZONTALLY across columns. Expect row totals:")
print(df.sum(axis=1))
print("\n" + "="*30 + "\n")
```

```python
# 4. The Drop Exception: axis=0
print("=== DROP WITH axis=0 ===")
print("Looking along the vertical axis to find and delete 'Student 0':")
print(df.drop('Student 0', axis=0))
print("\n" + "="*30 + "\n")
```

```python

# 5. The Drop Exception: axis=1
print("=== DROP WITH axis=1 ===")
print("Looking along the horizontal axis to find and delete 'Math (A)':")
print(df.drop('Math (A)', axis=1))
```
