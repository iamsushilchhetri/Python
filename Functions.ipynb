```python
import numpy as np
import matplotlib.pyplot as plt

mat_1 = np.linspace(1, 16, 100)

def sigmoid(matrix):
    """
    This is a sigmoid function
    This produces a series of floating numbers
    Plz only provide the integer or floating number
    Created Date : 02.17.26
    """
    return 1 / (1 + np.exp(-matrix))

y = sigmoid(mat_1)

plt.plot(mat_1, y)
plt.show()
```
