```python
# from a pandas dataframe single graph
batsman = pd.read_csv('D:\\Python\\sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# multiple graph
plt.plot(batsman['index'],batsman['V Kohli'],color = 'Red')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'Blue')
```

```python
# labels title
plt.title('Virat Kohli Vs Rohit Sharma IPP Runs comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.plot(batsman['index'],batsman['V Kohli'],color = 'Red')
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'Blue')
```

```python
plt.title('Comparison')
plt.plot(batsman['index'], batsman['V Kohli'],color = 'Red', marker = '.')
plt.plot(batsman['index'], batsman['RG Sharma'],color = 'Blue', marker = '.')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend --> location
plt.title('Comparison')
plt.plot(batsman['index'], batsman['V Kohli'],color = 'Red', marker = '.',label = 'Kohli')
plt.plot(batsman['index'], batsman['RG Sharma'],color = 'Blue', marker = '.', label = 'Rohit')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```

```python
# limiting axes
price = [50000,60000,70000,68000,63000,100000, 4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year, price)
plt.xlabel('Year')
plt.ylabel('Price')
plt.ylim(0,100000)
plt.xlim(2017,2021)
```

```python
# grid
price = [50000,60000,70000,68000,63000,100000, 45000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year, price)
plt.xlabel('Year')
plt.ylabel('Price')
plt.grid()
```

```python
# show
plt.title('Comparison')
plt.plot(batsman['index'], batsman['V Kohli'],color = 'Red', marker = '.',label = 'Kohli')
plt.plot(batsman['index'], batsman['RG Sharma'],color = 'Blue', marker = '.', label = 'Rohit')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.show()
```

# Bar Chart
# ..Bivariate Analysis
# ..Numerical vs categ
# ..use case--> Aggreate analysis of groups

```python
# bar chart
import matplotlib.pyplot as plt

# bar chart
children = [10, 20, 50, 39, 34]
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Pink']

# Swap the order: categories first, then numbers
plt.bar(colors, children)
```

```python
# horizonal bar
children = [10, 20, 50, 39, 34]
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Pink']

# Swap the order: categories first, then numbers
plt.barh(colors, children,color = 'Red')
```

```python
# color and label
df = pd.read_csv('D:\\Python\\batsman_season_record.csv')
df
```

```python
children = [10, 20, 50, 39, 34]
colors = ['RedRedRedRedRed', 'BlueBlueBlueBlue', 'GreenGreenGreenGreen', 'YellowYellowYellowYellow', 'PinkPinkPinkPink']

# Swap the order: categories first, then numbers
plt.bar(colors, children)
plt.xticks(rotation = 70)
```

```python
# Stacked Bar Chart
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
python_hours = [2, 3, 4, 3, 5]
ielts_hours = [3, 2, 1, 4, 2]

plt.bar(days, python_hours, label='Python')
plt.bar(days, ielts_hours, bottom=python_hours, label='IELTS')

plt.xlabel("Days")
plt.ylabel("Study Hours")
plt.title("Daily Study Hours")
plt.legend()

plt.show()
```
