```python
# Variables
name = "Sushil"
age = 22
height = 5.7
is_student = True

# Printing values
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", is_student)

# Checking data types
print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
```

```python
a = 10
b = 3

# Arithmetic Operators
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
print("Exponent:", a ** b)

# Comparison Operators
print(a > b)
print(a == b)

# Logical Operators
print(a > 5 and b < 5)
```

```python
number = 10

if number > 0:
    print("Positive number")
elif number == 0:
    print("Zero")
else:
    print("Negative number")
```

```python
# For loop
for i in range(5):
    print("For loop:", i)

# While loop
count = 0
while count < 5:
    print("While loop:", count)
    count += 1
```

```python
def greet(name):
    return f"Hello, {name}!"

def add(a, b):
    return a + b

print(greet("Sushil"))
print(add(5, 3))
```

```python
# List
numbers = [1, 2, 3, 4, 5]

# Accessing elements
print(numbers[0])

# List Comprehension
squares = [x**2 for x in numbers]
print("Squares:", squares)

# Even numbers using list comprehension
evens = [x for x in numbers if x % 2 == 0]
print("Even numbers:", evens)
```

```python
# Tuple (immutable)
coordinates = (10, 20)
print(coordinates)

# Set (unique values)
numbers = {1, 2, 2, 3, 4}
print(numbers)
```

```python
student = {
    "name": "Sushil",
    "age": 22,
    "course": "Computer Science"
}

print(student["name"])

# Looping through dictionary
for key, value in student.items():
    print(key, ":", value)
```

```python
# Lambda
square = lambda x: x**2
print(square(5))

numbers = [1, 2, 3, 4, 5]

# Map
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)

# Filter
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input!")
finally:
    print("Execution finished.")
```
