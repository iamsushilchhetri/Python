# 3D Line Plot

```python
import matplotlib.pyplot as plt

# 1. Set up the figure and tell Matplotlib we want a 3D plot
fig = plt.figure()
ax = plt.subplot(projection='3d')

# 2. Provide simple data points using basic lists
# Think of these as 5 specific points: (1,2,5), (2,4,4), (3,6,3), etc.
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [5, 4, 3, 2, 1]

# 3. Draw the line
# We'll add 'marker='o'' so you can clearly see the exactly where the 5 points are
# ax.scatter3D(x, y, z, marker='o', color='blue')
ax.plot3D(x, y, z, marker='o', color='blue')

# 4. Add labels so we know which axis is which
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# 5. Show the result
plt.show()
```

# 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,9))
ax = plt.subplot(projection='3d')

ax.plot_surface(xx, yy, z,cmap = 'viridis')
```

```python
# The "Sombrero" (Surface with Colormap)

fig = plt.figure(figsize=(10, 7))
ax = plt.subplot(111, projection='3d')

# Data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot with 'magma' colormap
surf = ax.plot_surface(X, Y, Z, cmap='magma', edgecolor='none', alpha=0.9)

# Add a color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_title('Visually Appealing Sombrero')
plt.show()
```

```python
# The "Ghostly" Wireframe (with Contours)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Use the same Z from above
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5, alpha=0.3)

# Project the "shadow" onto the bottom (z = -2)
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='coolwarm')

ax.set_zlim(-2, 2)
plt.show()
```

```python
#  The 3D "Galaxy" (Scatter Plot)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Generate 500 random points
n = 500
x = np.random.standard_normal(n)
y = np.random.standard_normal(n)
z = np.random.standard_normal(n)

# Color points by their distance from center
c = np.sqrt(x**2 + y**2 + z**2)

img = ax.scatter(x, y, z, c=c, cmap='YlGnBu', s=c*20, alpha=0.6)

fig.colorbar(img)
ax.set_title('3D Cluster Visualization')
plt.show()
```

# contour graph --> converts 3D graph to 

# contourf --- very very important

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup data
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)  # A simple wave pattern

# 2. Create the 2D plot
plt.figure(figsize=(8, 6))
# 'levels' defines how many color steps you want
cp = plt.contourf(X, Y, Z, levels=20, cmap='RdYlBu_r')

# 3. Add details
plt.colorbar(cp) # Shows the scale
plt.title('2D Filled Contour Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# We use the same X, Y, Z from above
# zdir='z' tells it to stack the contours along the vertical axis
# offset defines where on the Z-axis the plot sits
ax.contourf(X, Y, Z, zdir='z', offset=-1.5, cmap='viridis', alpha=0.5)
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.3) # Adding a faint surface for context

ax.set_zlim(-1.5, 1.5)
ax.set_title('3D Surface with Projected Contour Fill')

plt.show()
```
