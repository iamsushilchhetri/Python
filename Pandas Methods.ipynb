# apply (series + df)

```python
import pandas as pd

# Create a DataFrame with temperatures in Celsius
df = pd.DataFrame({
    'city': ['New York', 'London', 'Tokyo', 'Sydney'],
    'temp_c': [20, 15, 25, 30]
})

# Custom function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Apply the function to the temperature column
df['temp_f'] = df['temp_c'].apply(celsius_to_fahrenheit)
print(df)
```
