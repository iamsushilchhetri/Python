```python
class Student:
    # parameterized constructor
    def __init__(self, name , marks , sex):
        self.name = name
        self.marks = marks
        self.sex = sex

    def welcome(self):
        print(f'Welcome {self.name}, You are a M ale')

    def get_marks(self):
        return self.marks


s1 = Student('Sushil Chhetri', 97, 'Male')
s1.welcome()
print(s1.get_marks())
```
