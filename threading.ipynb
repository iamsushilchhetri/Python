```python
import threading
import time

def lights():
    print('Turn off the lights!!!! üí° ‚ùå')
    time.sleep(1)

def motor():
    print('Motor on gara')
    time.sleep(5)

def pasal():
    print('Bread, Butter, Eggs, Jam')
    time.sleep(1)

task_1 = threading.Thread(target=lights)
task_2 = threading.Thread(target=motor)
task_3 = threading.Thread(target=pasal)

task_1.start()
task_2.start()
task_3.start()

# Wait for all threads to finish
task_1.join()
task_2.join()
task_3.join()

print('All the chores are done ‚úÖ')
```

```python
# Print in Order

# üëâ Three threads:

# First prints "first"

# Second prints "second"

# Third prints "third" 
import threading
import time
def first():
    print('First ')

def second():
    print('Second ')

def third():
    print('Third ')

task_1 = threading.Thread(target = first)
task_1.start()
time.sleep(1)
task_1.join()
task_2 = threading.Thread(target = second)
task_2.start()
time.sleep(1)
task_2.join()
task_3 = threading.Thread(target = first)
task_3.start()
time.sleep(1)
task_3.join()
```
