```python
import numpy as np

# -----------------------------
# Step 1: Generate Proper Data
# -----------------------------

# Actual labels (must be 0 or 1)
actual = np.random.randint(0, 2, 90)

# Predicted probabilities (must be between 0 and 1)
predicted = np.random.rand(90)

print("Actual Labels:\n", actual)
print("-----------------------------")
print("Predicted Probabilities:\n", predicted)

# -----------------------------
# Step 2: Define BCE Function
# -----------------------------

def binary_cross_entropy(actual, predicted):
    epsilon = 1e-15  # prevents log(0)
    
    # Clip predictions to avoid log(0)
    predicted = np.clip(predicted, epsilon, 1 - epsilon)
    
    loss = -np.mean(
        actual * np.log(predicted) +
        (1 - actual) * np.log(1 - predicted)
    )
    
    return loss

# -----------------------------
# Step 3: Calculate Loss
# -----------------------------

loss = binary_cross_entropy(actual, predicted)

print("-----------------------------")
print("Binary Cross Entropy Loss:", loss)
```
